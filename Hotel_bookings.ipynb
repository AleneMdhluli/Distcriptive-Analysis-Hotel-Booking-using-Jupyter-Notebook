# Descriptive Statistic Analysis

# Hypothesis
1. More cancellations occur when prices are higher.


```python
# Lead time statistics 
lead_time_mean = df['lead_time'].mean()
lead_time_std = df['lead_time'].std()
lead_time_min = df['lead_time'].min()
lead_time_max = df['lead_time'].max()

print("\nLead Time Statistics:")
print(f"Mean lead time: {lead_time_mean:.2f}")
print(f"Standard deviation of lead time: {lead_time_std:.2f}")
print(f"Minimum lead time: {lead_time_min}")
print(f"Maximum lead time: {lead_time_max}")
```

```python
# Average daily Rate Statistics 
ADR_mean = df['adr'].mean()
ADR_std = df['adr'].std()
ADR_min = df['adr'].min()
ADR_max = df['adr'].max()


print ("\nAverage daily Rate Statistics:")
print(f"Mean Average daily Rate: {ADR_mean:.2f}")
print(f"Standard deviation of Average daily Rate: {ADR_std:.2f}")
print(f"Minimum Average daily Rate: {ADR_min}")
print(f"Maximum Average daily Rate: {ADR_max }")
```

```python
# Waiting list Statistics 
days_in_waiting_list_mean = df['days_in_waiting_list'].mean()
days_in_waiting_list_std = df['days_in_waiting_list'].std()
days_in_waiting_list_min = df['days_in_waiting_list'].min()
days_in_waiting_list_max = df['days_in_waiting_list'].max()


print("\nDays in Waiting List Statistics:")
print(f"Mean days in waiting list: {days_in_waiting_list_mean:.2f}")
print(f"Standard deviation: {days_in_waiting_list_std:.2f}")
print(f"Minimum days: {days_in_waiting_list_min}")
print(f"Maximum days: {days_in_waiting_list_max}")
```

```python
# Cancellation counts
cancellation_mean = df['is_canceled'].mean()
cancellation_std = df['is_canceled'].std()
canceled_min = df['is_canceled'].min()
canceled_max = df['is_canceled'].max()

print("\nCancellation counts Statistics:")
print(f"Mean Cancellation: {cancellation_mean:.2f}")
print(f"Standard Cancellation: {cancellation_std:.2f}")
print(f"Minimum Cancellation: {canceled_min}")
print(f"Maximum Cancellation: {canceled_max}")
```

# Visualisations 

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
print(cancelled_perc)

plt.figure(figsize = (5,4))
plt.title('Reservation status count')
plt.bar(x = ['Not canceled', 'Canceled'],
        height = df['is_canceled'].value_counts(),
        edgecolor = 'k',
        width = 0.7)


plt.show()
```

# Cancellation Statistics 
1. 62.8% of the reservations are not cancelled
2. 37.1% of the reservations are cancelled
3. This bar graph also depicts that more than half the reservations are cancelled which is a big problem.

```python
# Count reservations by hotel and cancellation status
reservation_counts = df.groupby(['hotel', 'is_canceled'])['is_canceled'].count().unstack(fill_value=0)

# Colors for canceled and not canceled reservations
colors = ['red', 'lightgreen'] 

# Bar plot for reservation status
reservation_counts.plot(kind='bar', stacked=True, color=colors)
plt.title('Reservation Status in Different Hotels', size=20)
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')
plt.xticks(rotation=15, ha='right') 
plt.legend(title='Status', labels=['Canceled', 'Not Canceled'])  
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Calculate cancellation rate per hotel
cancellation_rates = df.groupby('hotel')['is_canceled'].mean() * 100  

plt.figure(figsize=(8, 6))
plt.pie(cancellation_rates, labels=cancellation_rates.index, autopct="%1.1f%%", startangle=140)
plt.title("Cancellation Rate by Hotel (%)")
plt.axis('equal')  
plt.tight_layout()
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

# Cancellation rate based on hotel type statistics
1. City hotel recieves the highest highest reservations and cancellation rates.
3. City hotel recieves cancellation rates of 41.7% which is very high meaning that almost half the reservations are cancelling. 
2. Although Resort hotels might not recieve as much reservations as City hotels they are still able to keep majority of their
   reservations which is 72% who did not cancel.

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(15, 6)) 
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel', color='blue', linewidth=2)  
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel', color='orange', linewidth=2)
plt.title('Average Daily Rate in City and Resort Hotels', fontsize=16)  
plt.xlabel('Reservation Status Date', fontsize=12)
plt.ylabel('Average Daily Rate (ADR)', fontsize=12)  
plt.xticks(rotation=45, ha='right') 
plt.grid(True) 
plt.legend(fontsize=12)  
plt.tight_layout()  
plt.show()
```

# Average daily rate statistics 
1. The ADR flactuates over the years for both city and resort hotels. 
2. The prices spike for resort hotels alot during festive times especially around January of each year.
3. Putting both ADR's along each other, we find that the average daily rate for Resort hotels is far greater than that of City hotels.


```python
 #Create the count plot with customizations
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(15, 6))
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='coolwarm')
plt.title('Reservation Status per Month', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Reservations', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(['not cancelled', 'cancelled'])

if len(df['month'].unique()) > 5:
    plt.xticks(rotation=45, ha='right')

plt.tight_layout()  
plt.show()  
```

# Reservation status and cancellation statistics 
1. The most cancellations occur during January while the least occur during September.
2. Reservations increase during the summer months since these hotels are in the westren hemisphere. 
3. They are the lowest during December and January.
4. When cancellation is maximum, then our reservations are much less.

```python
data_per_month = df.groupby('month')['adr'].agg('sum').reset_index()  
plot_type = 'line'  
plot_func = plt.plot if plot_type == 'line' else plt.bar
plt.figure(figsize=(10, 6)) 
plot_func(data_per_month['month'], data_per_month['adr'], marker='o' if plot_type == 'line' else None, linestyle='-' if plot_type == 'line' else None)

plt.title('ADR per Month' if 'adr' in df.columns else 'Error: Column not found', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('ADR' if 'adr' in df.columns else 'Error', fontsize=12)
plt.xticks(rotation=45, ha='right')  
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
adr_per_month = df.groupby('month')['adr'].sum().reset_index()
plt.figure(figsize=(10, 6)) 
plt.scatter(adr_per_month['month'], adr_per_month['adr'])
plt.title('ADR per Month', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('ADR', fontsize=12)
plt.xticks(rotation=45, ha='right')  
plt.grid(True)  
plt.tight_layout() 
plt.show()
```

```python
reservations_per_month = df.groupby('month')['reservation_status'].count().reset_index()
reservations_per_month.rename(columns={'reservation_status': 'Count'}, inplace=True)  
adr_per_month = df.groupby('month')['adr'].mean().reset_index()

plt.figure(figsize=(12, 6))  
ax1 = plt.gca()  
plt.plot(reservations_per_month['month'], reservations_per_month['Count'], marker='o', linestyle='-', color='b', label='Number of Reservations')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Reservations', fontsize=12, color='b')
ax1.tick_params(axis='y', labelcolor='b')  
ax2 = ax1.twinx()  
plt.plot(adr_per_month['month'], adr_per_month['adr'], marker='s', linestyle='-', color='g', label='ADR')
plt.ylabel('ADR', fontsize=12, color='g')
ax2.tick_params(axis='y', labelcolor='g') 
plt.title('Reservations and ADR per Month', fontsize=14)
plt.xticks(rotation=45, ha='right')  
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Average Daily Rate and Reservation Statistics 
1. There is a positive corelation between the ADR and number of reservations.
2. The lower the ADR the more reservations the hotel recieves, however correlation does not mean causation. 
3. Festive seasons have the lowest ADR. We also see that the number of reservations depend on the time of the year or rather month of the year. 
4. Summer months recieve the most reservations and have the highest ADR. 
5. The increase in ADR may be as a result of high demand during the summer time. 

# Conclusion 
To sum up, the examination of hotel reservation and cancellation information reveals a number of important findings. With a higher cancellation rate (41.7%) than resort hotels (28%), city hotels have a greater difficulty. This highlights the significance of employing efficient reservation management techniques. Variations in the Average Daily Rate (ADR) throughout time, especially during holidays such as January, indicate how demand affects pricing dynamics. The month of January exhibits the highest cancellation rates, but the summer months observe peak reservations, which is correlated with elevated average daily fee. The necessity of pricing strategies that are in line with variations in demand is shown by the positive link found between ADR and reservation numbers. Through the utilisation of these insights, hotels can maximise revenue production while minimising cancellations by optimising pricing tactics and improving reservation management processes.
